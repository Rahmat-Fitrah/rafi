```python
data_mahasiswa = {} # Dictionary untuk menyimpan data mahasiswa, NIM sebagai key

def tambah_mahasiswa(): #coding untuk menambah mahasiswa
    """Menambahkan data mahasiswa baru."""
    print("\n--- Tambah Mahasiswa ---")
    nim = input("Masukkan NIM: ")
    if nim in data_mahasiswa:
        print("Error: Mahasiswa dengan NIM tersebut sudah ada!")
        return

    nama = input("Masukkan Nama: ")
    jurusan = input("Masukkan Jurusan: ")
    while True:
        try:
            ipk = float(input("Masukkan IPK: "))
            if not (0.0 <= ipk <= 4.0):
                raise ValueError
            break
        except ValueError:
            print("IPK tidak valid. Masukkan angka antara 0.0 dan 4.0.")

    data_mahasiswa[nim] = (nama, jurusan, ipk) # Menggunakan tuple untuk data mahasiswa
    print(f"Mahasiswa {nama} dengan NIM {nim} berhasil ditambahkan.")

def tampilkan_seluruh_mahasiswa():
    """Menampilkan seluruh data mahasiswa yang tersimpan."""
    print("\n--- Seluruh Data Mahasiswa ---")
    if not data_mahasiswa:
        print("Belum ada data mahasiswa.")
        return

    print("NIM\t\tNama\t\tJurusan\t\tIPK")
    print("-" * 50)
    for nim, data in data_mahasiswa.items():
        nama, jurusan, ipk = data
        print(f"{nim}\t{nama}\t\t{jurusan}\t\t{ipk:.2f}")
    print("-" * 50)

def cari_mahasiswa():
    """Mencari data mahasiswa berdasarkan NIM."""
    print("\n--- Cari Mahasiswa ---")
    nim_cari = input("Masukkan NIM mahasiswa yang ingin dicari: ")
    if nim_cari in data_mahasiswa:
        nama, jurusan, ipk = data_mahasiswa[nim_cari]
        print("\n--- Data Mahasiswa Ditemukan ---")
        print(f"NIM\t\t: {nim_cari}")
        print(f"Nama\t\t: {nama}")
        print(f"Jurusan\t\t: {jurusan}")
        print(f"IPK\t\t: {ipk:.2f}")
    else:
        print(f"Mahasiswa dengan NIM {nim_cari} tidak ditemukan.")

def update_ipk_mahasiswa():
    """Mengupdate data IPK mahasiswa berdasarkan NIM."""
    print("\n--- Update IPK Mahasiswa ---")
    nim_update = input("Masukkan NIM mahasiswa yang ingin diupdate IPK-nya: ")
    if nim_update in data_mahasiswa:
        nama_lama, jurusan_lama, ipk_lama = data_mahasiswa[nim_update]
        print(f"IPK lama untuk {nama_lama} (NIM: {nim_update}) adalah: {ipk_lama:.2f}")
        while True:
            try:
                ipk_baru = float(input("Masukkan IPK baru: "))
                if not (0.0 <= ipk_baru <= 4.0):
                    raise ValueError
                break
            except ValueError:
                print("IPK tidak valid. Masukkan angka antara 0.0 dan 4.0.")

        data_mahasiswa[nim_update] = (nama_lama, jurusan_lama, ipk_baru)
        print(f"IPK mahasiswa {nama_lama} (NIM: {nim_update}) berhasil diupdate menjadi {ipk_baru:.2f}.")
    else:
        print(f"Mahasiswa dengan NIM {nim_update} tidak ditemukan.")

def hapus_mahasiswa():
    """Menghapus data mahasiswa berdasarkan NIM."""
    print("\n--- Hapus Mahasiswa ---")
    nim_hapus = input("Masukkan NIM mahasiswa yang ingin dihapus: ")
    if nim_hapus in data_mahasiswa:
        nama_hapus, _, _ = data_mahasiswa.pop(nim_hapus)
        print(f"Mahasiswa {nama_hapus} dengan NIM {nim_hapus} berhasil dihapus.")
    else:
        print(f"Mahasiswa dengan NIM {nim_hapus} tidak ditemukan.")

def tampilkan_menu():
    """Menampilkan menu utama program."""
    print("\n--- Menu ---")
    print("1. Tambah Mahasiswa")
    print("2. Tampilkan Seluruh Mahasiswa")
    print("3. Cari Mahasiswa")
    print("4. Update IPK Mahasiswa")
    print("5. Hapus Mahasiswa")
    print("6. Keluar")

# Program utama
while True:
    tampilkan_menu()
    pilihan = input("Pilihan Anda: ")

    if pilihan == '1':
        tambah_mahasiswa()
    elif pilihan == '2':
        tampilkan_seluruh_mahasiswa()
    elif pilihan == '3':
        cari_mahasiswa()
    elif pilihan == '4':
        update_ipk_mahasiswa()
    elif pilihan == '5':
        hapus_mahasiswa()
    elif pilihan == '6':
        print("Terima kasih telah menggunakan program manajemen data mahasiswa.")
        break
    else:
        print("Pilihan tidak valid. Silakan masukkan angka antara 1 dan 6.")
```
