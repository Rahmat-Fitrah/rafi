```python
#improt libary matplotilip
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
try:
    df.info 
    df = pd.read_csv("personality_datasert.csv")
    data_dict = df.to_dict(orient='records')

    split_data = [
    row['Date;Close;High;Low;Open;Volume'].split(';')
    for row in data_dict
    ]

    columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
    df = pd.DataFrame(split_data, columns=columns)

    print(df)

except FileNotFoundError:
    print("file not found")
    
```

```python
plt.figure(figsize=(35,24))

for data in ['Close', 'High' , 'Low' , 'Open' , 'Volume']:
    plt.plot(df['Date'], df[data], marker='o', label=data)

plt.title("Tren Harga")
plt.ylabel("Percent")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

mean_value = np.mean(df["Close"])
print(mean_value)
plt.figure(figsize=(8, 5))
plt.hist(df["Close"], bins=15, color='skyblue', edgecolor='black', alpha=0.7, label='Data Distribusi')

# Add a horizontal line for the mean
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_value:.2f}')

# Add labels and legend
plt.title('Data Distribution width Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

#show the plot
plt.show()
```
