Buatlah Data latihan sederhana untuk klarifikasi data berdasarkan
<ol>
    <li>Usia : 18 - 60</li>
    <li>Transaksi perbulan : 0-50</li>
    <li>label : loyal(0), Tidak Loyal (1)</li>
    <li>Gender : P L </li>
</ol>

```python
import numpy as np
import pandas as pd

n_data = 100

usia = np.random.randint(18,60,size=n_data)
transaksi = np.random.randint(0,51, size=n_data)
label = np.random.randint(0,2, size=n_data)
gender = np.random.choice(['L','P'],size=n_data)

df = pd.DataFrame({
    'Usia' : usia,
    'Transaksi' : transaksi,
    'label' : label,
    'Gender' : gender
})

df.to_csv('dataLatih.csv',index=False)

df.head ()# untuk mengetahui 5 baris pertama dari isi dataframe
```
